```python
# logging gazebo model states
import rosbag
import matplotlib.pyplot as plt
bag = rosbag.Bag('2018-07-31-10-11-13.bag')
target_name="firefly"

xs=[]
ys=[]
zs=[]

for topic, msg, t in bag.read_messages(topics=['/gazebo/model_states']):
    idx=msg.name.index(target_name)
    xs.append(msg.pose[idx].position.x)
    ys.append(msg.pose[idx].position.y)
    zs.append(msg.pose[idx].position.z)
    
    
```

```python
# logging gazebo model states
import rosbag
import matplotlib.pyplot as plt
bag = rosbag.Bag('2018-07-31-10-16-54.bag')

xs=[]
ys=[]
zs=[]

for topic, msg, t in bag.read_messages(topics=['/firefly/command/trajectory']):
    xs.append(msg.points[0].transforms[0].translation.x)
    ys.append(msg.points[0].transforms[0].translation.y)
    zs.append(msg.points[0].transforms[0].translation.z)
```
